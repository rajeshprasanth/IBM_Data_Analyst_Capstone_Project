```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("popular-languages_new.csv")
```

```python
df_sorted= df.sort_values('Average Annual Salary',ascending=False)
```

```python
df_sorted.plot(kind='bar', x="Language", figsize=(20, 12))
plt.xlabel("Languages", size=15)
plt.ylabel("Average Annual Salary", size=15)
plt.title("Popular Languages", size=18)
```
